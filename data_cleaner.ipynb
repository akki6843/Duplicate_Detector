### Here the data cleaning takes place. Because of hardware restrictions only Tunics from samll dataset are used for this task.

here the target is to create a smaller dataset which contains information only related to Tunics

```python
import pandas as pd
import numpy as np
import cv2
```

```python
data = pd.read_csv("small-2oq-c1r.csv", error_bad_lines=False)
```

```python
data.head()
```

```python
pid = data['productId']
title = data['title']
description = data['description']
urls = data['imageUrlStr']
```

```python
urls_tun = []

for i, name in enumerate(list(pid)):
    try:
        if name[:3]=='TUN':
            urls_tun.append(urls[i])
            
    except:
        print("ignore")

```

```python
len(urls_tun)
```

```python
urls_200_tun = []
pid_tun = []
title_tun = []
for i in range(len(urls_tun)):
    a = urls_tun[i]
    b = a.split(";")
    c = str(b).split("/")
#     print(c[4])
    if c[4] == 'tunic':
        urls_200_tun.append(b[2])
        pid_tun.append(pid[i])
        title_tun.append(title[i])
```

```python
print(len(urls_200_tun))
print(len(pid_tun))
print(len(title_tun))
my_col = []
for i in zip(urls_200_tun, pid_tun, title_tun):
    my_col.append(i)
```

```python
E = pd.DataFrame(my_col,columns=["imageURL", "ProductId","title"])
# g = pd.concat([])
E.to_csv("tunic_urls.csv")
```
